```python
d = {"No": 0, "Yes":1} # bunun yerine label encoding de olurdu
```

```python
df["Attrition"] = df["Attrition"].map(d) # for un modern hali
```

```python
df['Attrition'].value_counts()
```

```python
bos = df[df["Attrition"].isnull()]
dolu = df[df["Attrition"].notnull()]
```

```python
bos.head()
```

```python
dolu.head()
```

```python
x = dolu.drop('Attrition', axis = 1)
y = dolu[['Attrition']]
```

```python
x.head()
```

```python
import matplotlib.pyplot as plt
```

```python
sns.countplot(x = df["EducationField"]);
plt.xticks(rotation = 90);
```

```python
x = pd.get_dummies(x,drop_first = True)
```

```python
x.head()
```

```python
x.shape
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
```

```python
g = GaussianNB()
b = BernoulliNB()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.2,random_state=42)
```

```python
g.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
g_pred = g.predict(x_test)
```

```python
accuracy_score(g_pred, y_test)
```

```python
confusion_matrix(g_pred, y_test)
```

```python
import seaborn as sns
```

```python
sns.heatmap(confusion_matrix(g_pred,y_test), annot = True)
```

```python
print(classification_report(g_pred, y_test))
```

```python
b.fit(x_train, y_train)
```

```python
b_pred = b.predict(x_test)
```

```python
accuracy_score(b_pred, y_test)
```

```python
confusion_matrix(b_pred, y_test)
```

```python
print(classification_report(b_pred, y_test))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
l = LogisticRegression()
```

```python
l.fit(x_train, y_train)
```

```python
l_pred = l.predict(x_test)
```

```python
accuracy_score(l_pred, y_test)
```

```python
confusion_matrix(l_pred, y_test)
```

```python
print(classification_report(l_pred, y_test))
```

```python
r.fit(x_train, y_train)
```

```python
r_pred = r.predict(x_test)
```
