```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Okuma ve ön işleme
data = pd.read_csv('LoansTrainingSet.csv')
data = data.drop(['Loan ID', 'Customer ID'], axis=1)
```

```python
# Kategorik değişkenlerin analizi
categorical_cols = ['Loan Status', 'Home Ownership', 'Purpose', 'Bankruptcies', 'Tax Liens']
for col in categorical_cols:
    print(f'Distribution of {col}:')
    value_counts = data[col].value_counts()
    print(value_counts)
    plt.figure(figsize=(8,6))
    ax = value_counts.plot(kind='bar')
    for i, v in enumerate(value_counts):
        ax.text(i, v, str(v), ha='center', va='bottom')
    plt.title(f'{col} Distribution')
    plt.show()
```

```python

# Sayısal değişkenlerin analizi
numerical_cols = ['Current Loan Amount','Term','Credit Score', 'Years in current job', 'Annual Income', 'Monthly Debt', 'Years of Credit History', 'Current Credit Balance', 'Maximum Open Credit']
for col in numerical_cols:
    print(f'Summary statistics for {col}:')
    print(data[col].describe())
    plt.figure(figsize=(8,6))
    sns.histplot(data[col])
    plt.title(f'{col} Distribution')
    plt.show()
```

```python
#convert 99999999 values in 'Current Loan Amount' to nan values

data["Current Loan Amount"] = data["Current Loan Amount"].replace(99999999, np.nan)

#fill nan values in "Current Loan Amount" with the best approtiate values to predict the most accurate values for loan status

data["Current Loan Amount"] = data["Current Loan Amount"].fillna(data["Current Loan Amount"].mean())
```

```python
col = "Current Loan Amount"
print(f'Summary statistics for {col}:')
print('number of Null values: ',  data["Current Loan Amount"].isnull().sum())
print(data[col].describe())
plt.figure(figsize=(8,6))
sns.histplot(data[col])
plt.title(f'{col} Distribution')
plt.show()
```

```python
#how can i fill null values of 'Current Loan Amount'

data["Current Loan Amount"].fillna(data["Current Loan Amount"].mean(), inplace=True)
data["Current Loan Amount"].isnull().sum()
data["Current Loan Amount"].value_counts()
```
