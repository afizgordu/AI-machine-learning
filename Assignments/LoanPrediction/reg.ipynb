```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Veri seti yükleme
data = pd.read_csv('LoansTrainingSet.csv')

# Regresyon modeli oluşturma
model = LinearRegression()
X = data[['Current Credit Balance', 'Maximum Open Credit', 'Monthly Debt', 'Years of Credit History']]
y = data['Current Loan Amount']
model.fit(X, y)

# Boş Current Loan Amount değerlerini doldurma
data['Current Loan Amount'] = data['Current Loan Amount'].fillna(model.predict(data[['Current Credit Balance', 'Maximum Open Credit', 'Monthly Debt', 'Years of Credit History']]))
```
