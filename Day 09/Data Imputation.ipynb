# Data Imputaiton

In this section I will learn how to impute missing datas.

```python
import pandas as pd
```

```python
df = pd.read_csv("ExampleData.csv")
```

```python
df.head()
```

```python
df.sample(3)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
imp = IterativeImputer(estimator=lr)
imputed_data = imp.fit_transform(df)
```

```python
pd.DataFrame(imputed_data).head()
```

```python
from sklearn.impute import KNNImputer
```

```python
knn = KNNImputer()
```

```python
imp_data = knn.fit_transform(df)
pd.DataFrame(imp_data,columns=[df.columns]).head()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(strategy='mean')
```

```python
imp_data = imp.fit_transform(df)
```

```python
pd.DataFrame(imp_data,columns=[df.columns]).head()
```
