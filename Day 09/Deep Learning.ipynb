```python
# DNN and Regression
```

```python
import pandas as pd
```

```python
df= pd.read_pickle("kc_house.pkl")
```

```python
df.head()
```

```python
x = df.drop('price', axis=1)
y = df[['price']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42, test_size=.2)
```

```python
model = Sequential()
model.add(Dense(120,activation= "relu"))
model.add(Dense(80,activation= "relu"))
model.add(Dense(64,activation= "relu"))
model.add(Dense(30,activation= "relu"))
model.add(Dense(20,activation= "relu"))
model.add(Dense(4,activation= "relu"))
model.add(Dense(1)) # for classification yes or no
model.compile(loss = "mse", optimizer = "adam")
```

```python
history = model.fit(x_train, y_train, validation_data= (x_test, y_test), batch_size=128, epochs=1500, verbose=0)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y_test,pred)
```

```python
mean_squared_error(y_test, pred)**.5
```

```python
loos_f = pd.DataFrame(history.history)
```

```python
loos_f.plot()
```
