# Computer Vision with Deep Learning

In this section I will learn Computer Vision with Deep Learning

```python
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.mnist # remember handwritten by pc
```

```python
(train_images, train_labels),(test_images, test_labels) = mnist.load_data()
```

```python
# normalize data
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
# Convolutional Neural Networks - CNN - Image Classification
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, InputLayer, Reshape, MaxPooling2D, Flatten 
```

```python
#build model

model = Sequential()
model.add(InputLayer(input_shape=(28,28))) #test_images.shape

model.add(Reshape(target_shape=(28, 28, 1))) # Add 1 more dimension for color channel

model.add(Conv2D(filters= 12, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(10)) # different 10 answers
model.compile(optimizer= 'adam', loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=["accuracy"])

```

```python
model.fit(train_images,train_labels,validation_split=.1,epochs=10,verbose=1)
```

```python
loss,accuracy = model.evaluate(test_images,test_labels)
```
